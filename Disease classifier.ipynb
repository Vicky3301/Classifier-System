```python
from sklearn.preprocessing import LabelEncoder
```

```python
lr=LabelEncoder()
```

```python
train["prognosis"]=lr.fit_transform(train["prognosis"])
test["prognosis"]=lr.fit_transform(test["prognosis"])
```

```python
pd.reset_option("display.max_columns")
pd.reset_option("display.max_rows")
```

```python
train=train.drop('Unnamed: 133',axis=1)
```

```python
train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
rfc=RandomForestClassifier()
lr=LogisticRegression()
sv=SVC()
knn=KNeighborsClassifier()
gb=GaussianNB()
xg=xgb.XGBClassifier()
```

```python
X_train=train.drop('prognosis',axis=1)
y_train=train['prognosis']
```

```python
X_test=test.drop('prognosis',axis=1)
y_test=test['prognosis']
```

```python
rfc.fit(X_train,y_train)
lr.fit(X_train,y_train)
sv.fit(X_train,y_train)
knn.fit(X_train,y_train)
gb.fit(X_train,y_train)
```

```python
xg.fit(X_train,y_train)
```

```python
rfc_pred=rfc.predict(X_test)
lr_pred=lr.predict(X_test)
sv_pred=sv.predict(X_test)
knn_pred=knn.predict(X_test)
gb_pred=gb.predict(X_test)
```

```python
xg_pred=xg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score ,mean_absolute_error,mean_squared_error  
from sklearn.metrics import classification_report, accuracy_score  
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score, matthews_corrcoef 
from sklearn.metrics import confusion_matrix
```

```python
print(accuracy_score(y_test,lr_pred))
print(accuracy_score(y_test,sv_pred))
print(accuracy_score(y_test,knn_pred))
print(accuracy_score(y_test,gb_pred))
print(accuracy_score(y_test,rfc_pred))
print(accuracy_score(y_test,xg_pred))
```

```python
print(precision_score(y_test,rfc_pred, average='macro'))
print(precision_score(y_test,lr_pred, average='macro'))
print(precision_score(y_test,sv_pred, average='macro'))
print(precision_score(y_test,knn_pred, average='macro'))
print(precision_score(y_test,gb_pred, average='macro'))
print(precision_score(y_test,xg_pred, average='macro'))
```

```python
models={"Logistic Regression":LogisticRegression(),"naive bays":GaussianNB(),
       "Support vector machine":SVC(),
       "K Neighbors Classifier":KNeighborsClassifier()}
```

```python
for name, model in models.items():
    print(f'Model name : {name}')
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    accuracy=accuracy_score(y_pred,y_test)
    precision=precision_score(y_pred,y_test,average='macro')
    print(f'accuracy   = {accuracy}')
    print(f'precision  = {precision}')
    print("------------------------------------")
```
