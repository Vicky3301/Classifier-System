```python
from sklearn.model_selection import train_test_split
```

```python
feature=df.drop("Class",axis=1)
```

```python
result=df["Class"]
```

```python
 X_train, X_test, y_train, y_test = train_test_split(feature, result, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
```

```python
rfc=RandomForestClassifier()
lr=LogisticRegression()
sv=SVC()
knn=KNeighborsClassifier()
gb=GaussianNB()
xg=xgb.XGBClassifier()
```

```python
rfc.fit(X_train,y_train)
lr.fit(X_train,y_train)
sv.fit(X_train,y_train)
knn.fit(X_train,y_train)
gb.fit(X_train,y_train)
xg.fit(X_train,y_train)
```

```python
rfc_pred=rfc.predict(X_test)
lr_pred=lr.predict(X_test)
sv_pred=sv.predict(X_test)
knn_pred=knn.predict(X_test)
gb_pred=gb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score ,mean_absolute_error,mean_squared_error  
```

```python
print(accuracy_score(y_test,rfc_pred))
print(accuracy_score(y_test,lr_pred))
print(accuracy_score(y_test,sv_pred))
print(accuracy_score(y_test,knn_pred))
print(accuracy_score(y_test,gb_pred))
```

```python
print(mean_absolute_error(y_test,rfc_pred))
print(mean_absolute_error(y_test,lr_pred))
print(mean_absolute_error(y_test,sv_pred))
print(mean_absolute_error(y_test,knn_pred))
print(mean_absolute_error(y_test,gb_pred))
```

```python
print(mean_squared_error(y_test,rfc_pred))
print(mean_squared_error(y_test,lr_pred))
print(mean_squared_error(y_test,sv_pred))
print(mean_squared_error(y_test,knn_pred))
print(mean_squared_error(y_test,gb_pred))
```

```python
from sklearn.metrics import classification_report, accuracy_score  
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score, matthews_corrcoef 
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,rfc_pred)
```

```python
print(precision_score(y_test,rfc_pred))
print(precision_score(y_test,lr_pred))
print(precision_score(y_test,sv_pred))
print(precision_score(y_test,knn_pred))
print(precision_score(y_test,gb_pred))
```

```python
print(f1_score(y_test,rfc_pred))
print(f1_score(y_test,lr_pred))
print(f1_score(y_test,sv_pred))
print(f1_score(y_test,knn_pred))
print(f1_score(y_test,gb_pred))
```

```python
print(recall_score(y_test,rfc_pred))
print(recall_score(y_test,lr_pred))
print(recall_score(y_test,sv_pred))
print(recall_score(y_test,knn_pred))
print(recall_score(y_test,gb_pred))
```
